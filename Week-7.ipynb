```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

n = 100
ps = [0.3, 0.6, 0.9]

colors = ['orange', 'cyan', 'pink']

for p, c in zip(ps, colors):
    x = np.arange(0, n+1)
    pmf = binom.pmf(x, n, p)
    mean, var = binom.stats(n, p)
    mode = int((n+1)*p)

    plt.figure(figsize=(8,4))
    plt.bar(x, pmf, alpha=0.7, color=c, edgecolor='black')
    plt.axvline(mean, color='red', linestyle='--', linewidth=2, label=f'Mean={mean:.1f}')
    plt.axvline(mode, color='gold', linestyle='-', linewidth=2, label=f'Mode={mode}')
    plt.axvline(mean+np.sqrt(var), color='green', linestyle=':', linewidth=2, label=f'StdDev=√{var:.1f}')
    plt.axvline(mean-np.sqrt(var), color='green', linestyle=':', linewidth=2)
    plt.title(f'Binomial Distribution (n={n}, p={p})', fontsize=14)
    plt.xlabel('On-time Deliveries', fontsize=12)
    plt.ylabel('Probability', fontsize=12)
    plt.legend()
    plt.show()
```
