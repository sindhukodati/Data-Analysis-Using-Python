```python
import numpy as np
import pandas as pd
```

```python
grades = np.array([67, 45, 88, 50, 39, 72, 54, 33, 90, 47])
threshold = 50
categories = np.where(grades >= threshold, 'Pass', 'Fail')
```

```python
grades_length = len(grades)
grades_dtype = grades.dtype
categories_dtype = pd.Categorical(categories).dtype
unique_categories = np.unique(categories)
```

```python
mean_grade = np.mean(grades)
median_grade = np.median(grades)
pass_rate = np.sum(categories == 'Pass') / len(grades) * 100
fail_count = np.sum(categories == 'Fail')
```

```python
def classify_grades(grades, threshold=50):
    return np.where(grades >= threshold, 'Pass', 'Fail')
```

```python
print(f"Length of grades vector: {grades_length}")
print(f"Data type of grades vector: {grades_dtype}")
print(f"Data type of categories vector: {categories_dtype}")
print(f"Unique categories: {unique_categories}")
print(f"Mean grade: {mean_grade}")
print(f"Median grade: {median_grade}")
print(f"Pass rate: {pass_rate}%")
print(f"Fail count: {fail_count}")
```
