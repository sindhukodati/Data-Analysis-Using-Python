```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
N_SHOTS = 10
P_SUCCESS_A = 0.7
P_SUCCESS_B = 0.5
N_GAMES = 100

simulated_successes_A = np.random.binomial(N_SHOTS, P_SUCCESS_A, N_GAMES)
k_values = np.arange(0, N_SHOTS + 1)
theoretical_pmf_A = binom.pmf(k_values, N_SHOTS, P_SUCCESS_A)

plt.figure(figsize=(14, 6))
print("\n--- 1. Simulation of Games  ---")

# Histogram (p=0.7)
plt.subplot(1, 2, 1)
plt.hist(simulated_successes_A, bins=np.arange(-0.5, N_SHOTS + 1.5, 1),
         density=True, color='skyblue', edgecolor='black', alpha=0.7,
         label=f'Simulation ({N_GAMES} Games)')
plt.plot(k_values, theoretical_pmf_A, 'ro-', linewidth=2, markersize=8,
         label=f'Theoretical PMF (p={P_SUCCESS_A})')
plt.title(f'Distribution of Successful Free Throws (p={P_SUCCESS_A})')
plt.xlabel('Number of Successful Baskets (k)')
plt.ylabel('Probability / Relative Frequency')
plt.xticks(k_values)
plt.legend()
plt.grid(axis='y', linestyle='--')


mean_A = N_SHOTS * P_SUCCESS_A
variance_A = N_SHOTS * P_SUCCESS_A * (1 - P_SUCCESS_A)
std_dev_A = np.sqrt(variance_A)
mode_A = k_values[np.argmax(theoretical_pmf_A)]
k_exact = 7
prob_exact_7 = binom.pmf(k_exact, N_SHOTS, P_SUCCESS_A)

prob_at_least_8 = binom.sf(7, N_SHOTS, P_SUCCESS_A) # P(X >= 8)
prob_fewer_than_5 = binom.cdf(4, N_SHOTS, P_SUCCESS_A) # P(X < 5)


simulated_successes_B = np.random.binomial(N_SHOTS, P_SUCCESS_B, N_GAMES)
theoretical_pmf_B = binom.pmf(k_values, N_SHOTS, P_SUCCESS_B)
mean_B = N_SHOTS * P_SUCCESS_B
variance_B = N_SHOTS * P_SUCCESS_B * (1 - P_SUCCESS_B)
std_dev_B = np.sqrt(variance_B)

# Histogram (p=0.5)
plt.subplot(1, 2, 2)
plt.hist(simulated_successes_B, bins=np.arange(-0.5, N_SHOTS + 1.5, 1),
         density=True, color='lightcoral', edgecolor='black', alpha=0.7,
         label=f'Simulation (p={P_SUCCESS_B})')
plt.plot(k_values, theoretical_pmf_B, 'bo-', linewidth=2, markersize=8,
         label=f'Theoretical PMF (p={P_SUCCESS_B})')
plt.title(f'Distribution of Successful Free Throws (p={P_SUCCESS_B})')
plt.xlabel('Number of Successful Baskets (k)')
plt.ylabel('Probability / Relative Frequency')
plt.xticks(k_values)
plt.legend()
plt.grid(axis='y', linestyle='--')

plt.tight_layout()
plt.show()

print("\n--- 2. Distribution Characteristics (p=0.7) ---")
print(f"Mean (E[X]): {mean_A:.2f}")
print(f"Variance (Var[X]): {variance_A:.2f}")
print(f"Standard Deviation (SD[X]): {std_dev_A:.2f}")
print(f"Most Likely Number (Mode): {mode_A}")

print("\n--- 3. Exact Probability (p=0.7) ---")
print(f"P(X=7): {prob_exact_7:.4f}")
print(f"Interpretation: This is the second-highest probability ({prob_exact_7:.2%}), confirming 7 as the expected value and center of the distribution.")

print("\n--- 4. Performance Consistency (p=0.7) ---")
print(f"P(X >= 8): {prob_at_least_8:.4f} (High performance)")
print(f"P(X < 5): {prob_fewer_than_5:.4f} (Poor performance)")
print("Reliability: The low P(X < 5) and high P(X >= 8) indicate high reliability; the player is highly unlikely to perform poorly.")

print("\n--- 5. What-if Scenario (p=0.7 vs p=0.5) ---")
print("Comparison:")
print(f"  Accuracy (p): {P_SUCCESS_A} -> {P_SUCCESS_B}")
print(f"  Mean (E[X]): {mean_A:.2f} -> {mean_B:.2f} (Drops significantly)")
print(f"  SD (Consistency): {std_dev_A:.2f} -> {std_dev_B:.2f} (Increases)")
print("Shape Change: The distribution shifts from left-skewed (p=0.7) to perfectly symmetrical (p=0.5).")
print("Consistency Effect: Lower accuracy (p=0.5) results in a wider, flatter distribution (higher SD), meaning **lower consistency** and less predictable outcomes.")
```
