```python
import numpy as np
import matplotlib.pyplot as plt

lambda_ = 15
hours_in_day = 8

call_volumes = np.random.poisson(lambda_, hours_in_day)

plt.figure(figsize=(10, 6))
plt.hist(call_volumes, bins=np.arange(min(call_volumes), max(call_volumes) + 1) - 0.5,
         edgecolor='darkblue', color='dodgerblue', alpha=0.75)

plt.title('Hourly Customer Call Volumes Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Number of Calls Received', fontsize=14)
plt.ylabel('Frequency (Number of Hours)', fontsize=14)

plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.xticks(np.arange(0, max(call_volumes) + 1, 2), fontsize=12)
plt.yticks(fontsize=12)

mean_calls = np.mean(call_volumes)
plt.axvline(mean_calls, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_calls:.2f}')
plt.legend()

plt.tight_layout()
plt.show()

print("Simulated Call Volumes for 8 Hours: ", call_volumes)

mean_calls = np.mean(call_volumes)
std_calls = np.std(call_volumes)
print(f"\nAnalysis of Call Volume Pattern:")
print(f"Mean Calls per Hour: {mean_calls:.2f}")
print(f"Standard Deviation: {std_calls:.2f}")

if std_calls > mean_calls * 0.3:
    variability_comment = "The pattern is highly variable, indicating unpredictable call volume."
else:
    variability_comment = "The pattern is relatively stable, with moderate variation in call volume."

print(f"\n{variability_comment}")
```
