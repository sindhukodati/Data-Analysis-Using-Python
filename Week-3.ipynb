```python
import kagglehub
import pandas as pd
import matplotlib.pyplot as plt

path = kagglehub.dataset_download("balaka18/email-spam-classification-dataset-csv")

print("Path to dataset files:", path)

df = pd.read_csv(path + '/emails.csv')

print("Columns in the dataset:", df.columns)

offer_counts = df['offer']
buy_counts = df['buy']
labels = df['category']

plt.figure(figsize=(10, 6))

plt.scatter(offer_counts[labels == 0], buy_counts[labels == 0], color='green', marker='^', label='Legitimate Emails')

plt.scatter(offer_counts[labels == 1], buy_counts[labels == 1], color='orange', marker='s', label='Spam Emails')

plt.title("Distribution of Offer & Buy Word Usage in Emails")
plt.xlabel("Count of 'offer' word")
plt.ylabel("Count of 'buy' word")

plt.legend()

plt.show()
```
